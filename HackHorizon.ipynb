```python
import pandas as pd

# Load file (ONLY THIS LINE)
df = pd.read_csv("HACKHORIZON 2.0 - 07-04-26 Updated Sheet.csv", header=1)

# Clean column names
df.columns = df.columns.str.strip()

clean_data = []

for _, row in df.iterrows():
    
    team_name = row['Team Name']
    college_name = row['College Name']

    members = [
        (row['Leader Name'], row['Leader Email']),
        (row['Member 2 Name'], row['Member 2 Email']),
        (row['Member 3 Name'], row['Member 3 Email']),
        (row['Member 4 Name'], row['Member 4 Email']),
        (row['Member 5 Name'], row['Member 5 Email'])
    ]

    for name, email in members:
        if pd.notna(name) and pd.notna(email):
            clean_data.append({
                "Name": name,
                "Email": email,
                "Team Name": team_name,
                "College Name": college_name
            })

new_df = pd.DataFrame(clean_data)

new_df.to_csv("FINAL_OUTPUT.csv", index=False)

print("✅ DONE bhai 🚀")
print("Total participants:", len(new_df))
```
